# 1. Basic Plots, Axis Labels and Titles

```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#pyplot Help:: https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
```

```python
x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)


plt.plot(x, y)
```

```python
# need to call plt.show() explicitly to diplay the plot
plt.show()
```

```python
#can also works with list
plt.plot([1,4,6,8], [3,8,3,5])
```

```python
#plot the chart
plt.plot(x,y)

#set x, y axis label and title
plt.xlabel("Current")
plt.ylabel("Volatage")
plt.title("Ohm's Law")

#define range of labels of axis
plt.xlim([20,80])
plt.ylim([200, 800])
plt.show()
```

```python
# Plotting dots instead of a continuous line
x1 = np.linspace(0, 10, 20)
y1 = x1 * 2

# Colour Blue, LineType : +
plt.plot(x1 ,y1, 'b+')

plt.xlabel("Current")
plt.ylabel("Volatage")
plt.title("Ohm's Law")

plt.show()
```

```python
#plotting multiple lines on same plot

a = np.linspace(0,5,10)
b = np.linspace(3, 6, 10)

#Plot mulitple plots
plt.plot(a,b, 'r-', a, b**2, 'b+', a, b**3, 'g*')
```

# 2. Figures and Subplots

```python
# you often need to create multiple plots in same figure.
# to create multiple plots, you can use plt.subplot(nrows, ncols, nsubplot)
```

```python
#2 graphs in 1 plot

c = np.linspace(1,10, 100)
d = np.log(c)

# initialize figure to draw subplots
plt.figure(1)

#create a subplot with 1 row and 2 columns
#create first subplot
plt.subplot(121)     #i.e. 1 Row,  2 Columns, 1st plot
plt.title("y = log(x)")
plt.plot(c, d)

#create second subplot
plt.subplot(122)      #i.e. 1 Row,  2 Columns, 2nd plot
plt.title("y = log(x)**2")
plt.plot(c, d**2)
plt.show()

```

```python
# 2 * 2 subplots in a plot

e = np.linspace(1,10, 100)

plt.figure(1)

#subplot 1: linear
plt.subplot(221)
plt.title("Linear")
plt.plot(e, e)

#subplot 2: Cubic
plt.subplot(222)
plt.title("Cubic")
plt.plot(e, e**2)

plt.figure(2)
#subplot 3: Log
plt.subplot(221)
plt.title("Log")
plt.plot(e,np.log(e))

#subplot 4: Exponential
plt.subplot(2,2,2)       # alternate way to specify plot number
plt.title("Exponential")
plt.plot(e, e**2)

plt.show()
```

# 3. Commonly Used Plots

```python
# 1. Boxplots
# 2. Histograms
# 3. Scatter Plot
# 4. Bar Plots
```

```python
#create dataframe

df = pd.read_csv("Global_Sales\market_fact.csv")
df.head()
```

# 3.1 BoxPlot

```python
#Boxplot for Order_Quantity
plt.boxplot(df['Order_Quantity'])
plt.show()
```

```python
# Box plot for Sales

plt.boxplot(df['Sales'])
plt.show()
```

```python
# as we can see, plot is unreadable, since sales varies across wide range
df["Sales"].describe()
```

```python
# here, max value and 75% quantile has a lot of difference. 
# Solution is to change scale of X-Axis and convert it into Logarithmic Scale

plt.subplot(1,2,1)
plt.boxplot(df['Sales'])

plt.subplot(1,2,2)
plt.boxplot(df['Sales'])
plt.yscale('log')

plt.show()


```

# 3.2 Histogram

```python
plt.hist(df['Sales'])
plt.show()
```

```python
# histogram can be more readable on logarithmic scale
plt.hist(df['Sales'])
plt.yscale('log')
plt.show()
```
