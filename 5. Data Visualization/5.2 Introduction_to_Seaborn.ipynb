# 1. Visualising Univariate Distributions

```python
# we have visualised univariate distributions using Matplotlib. Now we can try same using Seaborn
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global_Sales/market_fact.csv")

sns.set_style("whitegrid")
```

# Histograms and Density Plots

```python
# Simple Density PLot

sns.distplot(df["Shipping_Cost"])
plt.show()
```

```python
# You can also plot called as Rug PLot which plots actual data points as small vertical bars. 
# It is specified as argument to distplot

sns.distplot(df["Shipping_Cost"], rug=True)
plt.show()
```

```python
# Simple Desity plot can be created by using hist = false
sns.distplot(df["Shipping_Cost"], hist=False)
plt.show()
```

```python
# As Seaborn is based on matplotlib, usual matplotlib functions work well with seaborn.
# PLot multiple graphs using subplot
```

```python
plt.figure(1)

# Sales
plt.subplot(2,2,1)
plt.title("Sales")
sns.distplot(df["Sales"])

#Profit
plt.subplot(2,2,2)
plt.title("Profit")
sns.distplot(df["Profit"])

# Order_quantity
plt.subplot(2,2,3)
plt.title("Order_Quantity")
sns.distplot(df["Order_Quantity"])

#Shipping Cost
plt.subplot(2,2,4)
plt.title("Shipping Cost")
sns.distplot(df["Shipping_Cost"])

plt.show()
```

# Boxplot

```python
sns.boxplot(df["Order_Quantity"])
plt.title("Order Quantity")
plt.show()
```

```python
# Above boxplot is spread over x-axis. To make it vertical, set it as y
sns.boxplot(y=df["Order_Quantity"])
plt.show()
```

# 2.2 Visualizing Bivariate distributions

```python
sns.jointplot(df["Sales"], df["Profit"])
plt.show()
```

```python
# plotting same plot using filtered df

df1 = df[(df.Profit < 10000) & (df.Sales < 20000)]
sns.jointplot("Sales", "Profit", df1)  #alternate way to plot

plt.show()
```

```python
# You can adjust arguments of jointplot() to make more readable plot. e.g. specify kind = hexto create hexbin plot 

df2 = df[(df.Profit < 100) & (df.Profit > -100) & (df.Sales < 200)]
sns.jointplot(df2.Sales, df2.Profit, kind='hex', color='b' )
plt.show()
```

# Plotting pairwise relationships

```python
btc = pd.read_csv("Crypto/bitcoin_price.csv")
ether = pd.read_csv("Crypto/ethereum_price.csv")
ltc = pd.read_csv("Crypto/litecoin_price.csv")
monero = pd.read_csv("Crypto/monero_price.csv")
neo = pd.read_csv("Crypto/neo_price.csv")
quantum = pd.read_csv("Crypto/qtum_price.csv")
ripple = pd.read_csv("Crypto/ripple_price.csv")
 
print(btc.columns)

# as columns in all Dataframes are same, we suffix some identifier
btc.columns = btc.columns.map(lambda x: str(x) + '_btc')
ether.columns = ether.columns.map(lambda x: str(x) + '_et')
ltc.columns = ltc.columns.map(lambda x: str(x) + '_ltc')
monero.columns = monero.columns.map(lambda x: str(x) + '_mon')
neo.columns = neo.columns.map(lambda x: str(x) + '_neo')
quantum.columns = quantum.columns.map(lambda x: str(x) + '_qt')
ripple.columns = ripple.columns.map(lambda x: str(x) + '_rip')

btc.head()  
```

```python
#merge all files by date

m1 = pd.merge(btc, ether, how="inner", left_on="Date_btc", right_on="Date_et")
m2 = pd.merge(m1, ltc, how="inner", left_on="Date_btc", right_on="Date_ltc")
m3 = pd.merge(m2, monero, how="inner", left_on="Date_btc", right_on="Date_mon")
m4 = pd.merge(m3, neo, how="inner", left_on="Date_btc", right_on="Date_neo")
m5 = pd.merge(m4, quantum, how="inner", left_on="Date_btc", right_on="Date_qt")
crypto = pd.merge(m5, ripple, how="inner", left_on="Date_btc", right_on="Date_rip")

crypto.head()
```

```python
#Subsetting only close prices column for plotting
curr = crypto[["Close_btc","Close_et","Close_ltc","Close_mon","Close_neo","Close_qt", "Close_rip"]]
curr.head()
```

```python
sns.pairplot(curr)
plt.show()
```

```python
cor = curr.corr()
round(cor, 3)
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(cor, cmap='YlGnBu', annot=True)
plt.show()
```
