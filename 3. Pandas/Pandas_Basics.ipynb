# 3.2 Pandas DataFrame

```python
rNums = np.random.randint(low = 20, high = 100, size=[20,])
rNames = np.random.choice(['Ajay', "Ranbir", 'Hritik', 'Anil'], 20)
rVals = np.random.choice([10,11,12,13,14], 20)

print(type(rNums))

#Zip function joins two/more tuples together. i.e. it maps 1st element of each tuple and creates a row for it
d = list(zip(rNums, rNames, rVals))

#for value in d:
#    print(value)
    
# create pandas data frame
df = pd.DataFrame(data= d, columns=['RNUM', 'RNAME', 'RVAL'])
print("DataFrame df = \n", df, '\n')

print("type(df) = ", type(df), '\n')

#Creating dataframe using dictionary. Specify column names along with data
df2 = pd.DataFrame({'RNUM':rNums, 'RNAME':rNames, 'RVAL':rVals})
print('DF2:: df2\n', df2, '\n')
```

# Merging and concatting dataframes

```python

d1 = pd.DataFrame([['A',1], ['B', 2]], columns=['col1', 'number'])
d2 = pd.DataFrame([['C',3, 'Lion'], ['D', 4, 'Tiger']], columns=['letter', 'number', 'animal'])
```

```python
d1
```

```python
d2
```

```python
# if ignore_index is not used/False, concat will also append index column of each individual DF and this will repeat indices.
#axis = 0 means concat row-wise
pd.concat([d1, d2], sort=False, axis = 0, ignore_index=True)
```

```python
#concat column-wise
pd.concat([d1, d2], sort=False, axis = 1)
```

```python
#Joining 2 dataframes
d3 = pd.DataFrame({ "city":['Kolhapur', 'Pune', 'Nashik', 'Mumbai'],
                    "temperature" : [34, 38, 32, 33]})

d4 = pd.DataFrame({ "city":['Kolhapur', 'Pune', 'Nashik'],
                    "humidity" : [68, 65, 75]})
```

```python
#this will perform inner join on both dataframes showing only common values
d5 = pd.merge(d3, d4, on='city')
d5
```

```python
#this will perform outer join on both dataframes showing all values
# how  parameter can have values as outer, left, right and inner.(default: inner)
d6 = pd.merge(d3, d4, on='city', how="outer")
d6
```

```python
#adding extra column in DF

#first define function which will be applied on values of DF
def hot_temp(x):
    return x > 30

wdf['HotDay'] = wdf['temperature'].apply(hot_temp)
wdf
```

```python
#This extra column can also be added using lambda functions
wdf["LambdaHotDay"] = wdf['temperature'].apply(lambda x: x>30 )
wdf
```
